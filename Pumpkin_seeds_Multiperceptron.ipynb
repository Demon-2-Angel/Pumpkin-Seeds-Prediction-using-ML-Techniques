```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import math
from sklearn.preprocessing import MinMaxScaler
```

```python
y_map = {label: value for value, label in enumerate(np.unique(y))}
y_labeled = y.map(y_map).values
print(y_labeled)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
X_train = scaler.fit_transform(np.array(X_train).reshape(-1,1))
y_train = scaler.transform(np.array(y_labeled).reshape(-1,1))
```

```python
window =1
model = Sequential()
model.add(Dense(40, input_dim=window, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1))
```

```python
import keras
opt  = keras.optimizers.Adagrad(learning_rate = 0.05)
```

```python
model.compile(optimizer=opt ,loss='mean_squared_error')
model.summary()
```

```python
import tensorflow as tf

X_train = tf.stack(X_train)
y_train = tf.stack(y_train)
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=10)
```
